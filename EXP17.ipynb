```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
display(df)
```

```python
#Line Chart
plt.plot(df['Days'],df['Study_Hours'],marker='o')        #days on x axis and study hours on y axis marker is additional for design as where graph changes direction a dot is marked there
plt.xlabel("Days")
plt.ylabel("Study Hours")
plt.title("Siddhant")
plt.show()
```

```python
#Advanced Line Graph
plt.figure(figsize=(7,4))            #width 7 inch height 4 inch

plt.plot(df['Days'],df['Study_Hours'],marker='o',color='green',label='Study Hours')

plt.plot(df['Days'],df['Marks'],marker='s',color='red',label='Marks')

plt.title("Study Hours and Marks Trend Siddhant")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#Bar Chart (Comparison)

plt.bar(df['Days'],df['Marks'],color='yellow',edgecolor='black')
plt.grid(axis='y',linestyle='--',alpha=0.5)         #alpha is used to define intensity
plt.xlabel("Days")
plt.ylabel("Marks")
plt.title("Siddhant")
plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Days'],df['Marks'],color='red')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, y, str(y),
           ha='center',va='bottom')
plt.grid(axis='y')            #adds horizontal grid
plt.xlabel("Days")
plt.ylabel("Marks")
plt.title("Siddhant")
plt.show()
```

```python
import numpy as np

x=np.arange(len(df['Days']))      #position of bars
width = 0.36                       #width of each bar
plt.bar(x - width/2, df['Study_Hours'],width, label='Study Hours')
plt.bar(x + width/2, df['Marks'],width, label='Marks')
plt.xticks(x, df['Days'])                     # used to print days on x axis try commenting this line to see the difference
plt.title("Study Hours VS Marks Siddhant")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Example data(more values for better histogram)
marks=[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
#Calculate mean
mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks, bins=3, color='skyblue',edgecolor='black')
#Mean line
plt.axvline(mean_value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title("Siddhant")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
#Scatter Plot
plt.scatter(df['Study_Hours'],df['Marks'])
plt.xlabel("Days")
plt.ylabel("Marks")
plt.title("Siddhant")
plt.show()
```

```python
# Conditional Scatter Plot
#Example category
df['Result'] = ['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'],c=colors)
plt.xlabel("Days")
plt.ylabel("Marks")
plt.title("Siddhant")
plt.grid(axis='y')
plt.grid(axis='x')
plt.show()
```

```python
#Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}
df = pd.DataFrame(data)
print(df)
```

```python
#Line Plot
sns.lineplot(x='Day',y='Sales',data=df,errorbar=None)
plt.title("Seaborn Line Plot Siddhant")
plt.show()
```

```python
#Bar Plot
sns.barplot(x='Day',y='Sales',data=df,errorbar=None)   #try without errorbar and see why is it used
plt.title("Seaborn Bar Plot Siddhant")
plt.show()
```

```python
#Histogram
sns.histplot(df['Sales'],bins=5)
plt.title("Seaborn Histogram Siddhant")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title("Scatter Plot Siddhant")
plt.show()
```
